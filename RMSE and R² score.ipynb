```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Data = pd.read_csv("./onlinefraud.xls")
Data.head()
```

```python
data = Data.sample(n=500, random_state=42)
data.head()
```

```python
len(data)
```

```python
data["type"] = data["type"].map({"CASH_OUT": 1, "PAYMENT": 2, "CASH_IN": 3, "TRANSFER": 4, "DEBIT":5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
data.head()
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data[["isFraud"]])
```

```python
from sklearn.svm import SVC
Xtrain, Xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30, random_state=50)
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer to replace NaNs with the mean (or median, or most frequent)
imputer = SimpleImputer(strategy='mean')
xtrain = imputer.fit_transform(Xtrain)
xtest = imputer.transform(Xtest)
```

```python
model_li =  SVC(kernel = 'linear', random_state = 0)
model_li.fit(xtrain, ytrain)
model_li.score(xtest, ytest)
```

```python
features = np.array([[4, 9000.60, 9000.60, 0.0]])
print(model_li.predict(features))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred = model_li.predict(xtest)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
le = LabelEncoder()
ytest_num = le.fit_transform(ytest)
ypred_num = le.transform(y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(ytest_num, ypred_num))
r2 = r2_score(ytest_num, ypred_num)
```

```python
print(f"\nRMSE (numeric comparison only): {rmse:.3f}")
print(f"R² Score (numeric comparison only): {r2:.3f}")
```
